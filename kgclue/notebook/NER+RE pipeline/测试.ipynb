```python
import sys,os
import pandas as pd
import random
import json
import numpy as np
from torch.utils.data import DataLoader
from nlp_basictasks.tasks import cls
from collections import defaultdict,Counter
from tqdm import tqdm
```

```python
test_data=[]
with open("../../qa_data/private_test.json") as f:
    lines=f.readlines()
    for line in lines:
        test_data.append(json.loads(line))
```

# 导入模型

```python
cls_model=cls(model_path='/home/xhsun/Desktop/KG/kgCLUE/code/save_dir/',device='cuda',is_finetune=True)
```

```python
for i,example in enumerate(test_data):
    question=example['question']
    predict_probs=cls_model.predict(is_pairs=False,dataloader=[question])
    id2label={id_:label for label,id_ in cls_model.label2id.items()}
    predict_tags=[id2label[id_] for id_ in np.argmax(predict_probs,axis=1)][0]
    test_data[i].update({"relation":predict_tags})
```

# 读取知识图谱，获取sp_map

```python
sub_map = defaultdict(list)
obj_map = defaultdict(list)
so_map=defaultdict(list)
sp_map=defaultdict(list)
kg_path='/home/xhsun/Desktop/KG/kgCLUE/knowledge/Knowledge.txt'
with open(kg_path) as f:
    lines=f.readlines()
    
entities_set=set()
relations_set=set()
print(len(lines))

#new_triples=[]
for i in tqdm(range(len(lines))):
    line=lines[i]
    l = line.strip().split('\t')
    s = l[0].strip()
    p = l[1].strip()
    o = l[2].strip()
    if s=='' or p=='' or o=='':
        print('-'*100)
        continue
    entities_set.add(s)
    entities_set.add(o)
    relations_set.add(p)
    #new_triples.append((s,p,o))
    sub_map[s].append((p, o))
    obj_map[o].append((s, p))
    so_map[(s,o)].append(p)
    sp_map[(s,p)].append(o)
```

```python
print('大规模图谱中一共有{}个三元组，{}个实体，{}个关系'.format(len(lines),len(entities_set),len(relations_set)))
```

```python
def text_match(attribute_list,sentence):
    idx = -1
    for i,attribute in enumerate(attribute_list):
        if attribute in sentence:
            idx = i
            break
    if -1 != idx:
        return attribute_list[idx]
    else:
        return ""
    
submit_examples=[]
for example in test_data:
    id_,question,topic_entity,relation=example['id'],example['question'],example['topic_entity'],example['relation']
    #assert relation in relations_set
    if topic_entity not in entities_set:
        print(example)
    all_relation_for_current_entity=sub_map[topic_entity]#存在实体歧义的问题，暂时无法考虑
    if all_relation_for_current_entity!=[]:
        all_relation_for_current_entity=[p for p,o in all_relation_for_current_entity]
    
    r=text_match(attribute_list=all_relation_for_current_entity,sentence=question)
    if r!='':
        relation=r
        tail=sp_map[(topic_entity,relation)][0]#关系存在语义相似的问题
    else:
        #文字匹配不上
        tail=sp_map[(topic_entity,relation)]#关系存在语义相似的问题
        if tail==[]:
            #预测的关系也匹配不上
            for each_p in all_relation_for_current_entity:
                a=set(list(each_p))
                b=set(list(relation))
                if len(a.intersection(b))!=0:
                    relation=each_p
                    break
            tail=sp_map[(topic_entity,relation)]#关系存在语义相似的问题
        if tail==[]:
            tail="UNK"
        else:
            tail=tail[0]
    ans=str(topic_entity)+" ||| "+str(relation)+" ||| "+str(tail)
    submit_examples.append({"id":id_,"answer":ans})
```
